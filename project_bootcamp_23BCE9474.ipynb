```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#First we have to create a data frame using pandas.
#here heart_failure_clinical_records data is taken to prepare data frame. 
df = pd.read_csv('heart_failure_clinical_records.csv')
df
#Here we plot a graph between age and platelets.
fig = plt.figure()
plt = fig.add_axes([0.1,0.1,1,1])
x=np.array(df['age'])
x.sort()
y=np.array(df['platelets'])
y.sort()
plt.plot(x,y,'r')
plt.set_xlabel("age")
plt.set_ylabel("platelets")
```
